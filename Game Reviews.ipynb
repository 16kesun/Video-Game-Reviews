```python
df_Xone = df.loc[df['Platform'] == 'XOne']
summary_lst = ['Global_Sales','Critic_Score','User_Count']
empty_lst = []
for feature in summary_lst:
    empty_lst.append(df_Xone[feature].mean())
dictionary = dict(zip(summary_lst,empty_lst))
dictionary
```

```python
def get_console(console):
    df_console = df.loc[df['Platform'] == console]
    summary_lst = ['Global_Sales','Critic_Score','User_Count']
    empty_lst = []
    for feature in summary_lst:
        empty_lst.append(df_console[feature].mean())
    dictionary = dict(zip(summary_lst,empty_lst))
    return dictionary
        
```
